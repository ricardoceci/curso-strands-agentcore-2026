# Lab 2 — Múltiples Proveedores de Modelos

Una de las ventajas clave de Strands es que **desacopla tu código del proveedor de LLM**. En este lab conectamos el mismo `Agent` a tres proveedores distintos:

| Proveedor | Modelo | Requiere |
|-----------|--------|----------|
| OpenAI | GPT-4o Mini | `OPENAI_API_KEY` en `.env` |
| Anthropic | Claude Sonnet 4.6 | `ANTHROPIC_API_KEY` en `.env` |
| Ollama | Qwen3 4B | Ollama instalado localmente |

La interfaz del agente es idéntica en los tres casos.

````python
from strands import Agent
from strands.models.openai import OpenAIModel
from dotenv import load_dotenv
import os
load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")


````

## OpenAI — GPT-4o Mini

`OpenAIModel` recibe el `model_id` y los argumentos del cliente de OpenAI. El agente resultante tiene exactamente la misma interfaz que en Lab 1: no cambia nada en cómo lo usás.

````python
model = OpenAIModel(model_id="gpt-4o-mini", client_args={
    "api_key": OPENAI_API_KEY
})

agente = Agent(model=model)
````

Probamos que el agente responde correctamente usando el modelo de OpenAI.

````python
agente("Quien sos?")
````

## Anthropic — Claude Sonnet 4.6 (sin Bedrock)

Aquí usamos la **API de Anthropic directamente** (no a través de Bedrock). La diferencia clave: `max_tokens` es **obligatorio** en el cliente nativo de Anthropic.

> Usamos el mismo modelo que Bedrock pero con credenciales distintas.

````python
from strands import Agent
from strands.models.anthropic import AnthropicModel

from dotenv import load_dotenv
import os
load_dotenv()

ANTHROPIC_API_KEY = os.getenv("ANTHROPIC_API_KEY")

model = AnthropicModel(model_id="claude-sonnet-4-6", client_args={
    "api_key": ANTHROPIC_API_KEY
},max_tokens=400)

agent = Agent(model=model)

agent("Quien sos?")
````

## Ollama — Modelo Local (sin costo, sin internet)

[Ollama](https://ollama.com/) permite correr modelos open-source localmente. Con `OllamaModel` apuntamos al servidor local (por default en `localhost:11434`).

> **Prerequisito:** tener Ollama instalado y el modelo descargado:
> ```bash
> ollama pull qwen3:4b
> ```

Notar que `outputTokens` es muy alto (1336) porque Qwen3 incluye su razonamiento interno en la respuesta por default.

````python
from strands.models.ollama import OllamaModel

model = OllamaModel(model_id="qwen3:4b", host="http://localhost:11434")

agente_ollama = Agent(model=model)

agente_ollama("Quien sos?")
````
