# Lab 1 — Tu primer Agente con Strands

En este lab creamos el agente más simple posible con Strands Agents.

Con una sola línea instanciamos el agente y con otra llamada iniciamos la conversación. Por default, `Agent()` usa **Amazon Bedrock** como proveedor (Claude Sonnet 4.6).

---

```python
from strands import Agent

agente = Agent()

agente("Hola quien sos?")
```

## Historial de conversación

Cada `Agent` mantiene automáticamente un historial en `agente.messages`. Esta lista sigue el formato estándar de la API: turnos alternados `user` / `assistant`.

Junto con el texto, cada mensaje incluye métricas de uso (`inputTokens`, `outputTokens`) y de latencia (`latencyMs`, `timeToFirstByteMs`).

```python
agente.messages
```

## Conversación multi-turno

Al llamar al agente nuevamente, Strands incluye **todo el historial previo** en el contexto. Observá cómo `inputTokens` crece en cada llamada: el contexto acumulado se envía completo al modelo en cada turno.

```python
agente("Quien es tu creador?")
```

Inspeccionamos el historial actualizado. Ahora tiene 4 mensajes: los 2 del primer turno más los 2 del segundo.

```python
agente.messages
```

## Inspeccionando el modelo

Podemos ver qué modelo está usando el agente con `agente.model`. Strands abstrae el proveedor detrás de una interfaz uniforme: cambiamos de proveedor sin tocar el resto del código.

```python
agente.model
```

## Cambiando el System Prompt

El `system_prompt` define el rol y las instrucciones base del agente. Podemos modificarlo en cualquier momento; la nueva instrucción se aplica desde el siguiente turno.

> **Nota:** el historial de mensajes anteriores se preserva. El modelo ve tanto el nuevo system prompt como toda la conversación previa.

```python
agente.system_prompt = "Sos un asistente experto en Soccer"
```

```python
agente("Quien sos?")
```

## Contando mensajes

Cada turno de conversación agrega 2 mensajes al historial (uno de `user`, uno de `assistant`). Después de 3 turnos tenemos 6 mensajes en total.

```python
print("Cantidad de mensajes:", len(agente.messages))
```

## Activando logs de Strands

Strands usa el módulo estándar `logging` de Python. Habilitando el nivel `DEBUG` podemos ver exactamente qué request se envía al modelo: historial completo, system prompt y parámetros de inferencia.

Esto es muy útil para depurar el comportamiento del agente.

```python
import logging

logging.getLogger("strands").setLevel(logging.DEBUG)

logging.basicConfig(
    format="%(levelname)s | %(name)s | %(message)s",
    handlers=[logging.StreamHandler()]
)
```

Con los logs activados ejecutamos otra pregunta. En la salida vas a ver el request completo que Strands envía a Bedrock, incluyendo todos los mensajes del historial.

```python
agente("Cual es el mejor jugador de soccer del mundo?")
```
